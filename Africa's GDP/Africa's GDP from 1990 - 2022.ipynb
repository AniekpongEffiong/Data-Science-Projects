```python
# importing all modules
from sjvisualizer import DataHandler
from sjvisualizer import Canvas
from sjvisualizer import StackedBarChart
from sjvisualizer import BarRace
from sjvisualizer import PieRace
from sjvisualizer import StaticText
from sjvisualizer import Date
from sjvisualizer import Total
from sjvisualizer import StaticImage
import json
```

```python
# define constants
FPS = 60
DURATION = 0.25
BACKGROUND_COLOR = (255, 255, 255)
TITLE_COLOR = (238,177,132)
FONT_COLOR = (0,0,0)
BAR_COLOR = (255,255,255)
STACK_COLOR = (255,163,42)
PIE_COLOR = (255,163,42)
NEW_COLOR = (124,70,32)
```

```python
# load the data from the excel file
df = DataHandler.DataHandler(excel_file="data/GDP.xlsx", number_of_frames=FPS*DURATION*60).df
```

```python
# loading colors
with open("stacks.json") as f:
    stacks = json.load(f)
```

```python
# loading colors
with open("color.json") as f:
    color = json.load(f)
```

```python
# creating the canvas to which we want to draw all elements
canvas = Canvas.canvas(bg=STACK_COLOR)
```

```python
# drawing a rectangle around working area
canvas.canvas.create_rectangle(-1,-1,1771,1771)
```

```python
# creating the bar race
bar_race = BarRace.bar_race(canvas=canvas.canvas, df=df, colors=color, width=770, height=770, x_pos=470, y_pos=200, shift=200, unit = "$ ", font_color=FONT_COLOR, number_of_bars=12)
canvas.add_sub_plot(bar_race)
```

```python
# adding an additional chart type
pie_race = PieRace.pie_plot(canvas=canvas.canvas, df=df, colors=color, back_ground_color=PIE_COLOR, font_color=STACK_COLOR, width=350, height=350, x_pos=1045, y_pos=500)
canvas.add_sub_plot(pie_race)
```

```python
# creating the stacked bar chart
stacked = StackedBarChart.stacked_bar_chart(canvas=canvas.canvas, df=df, x_pos=1270, y_pos=40, width=100, height=100, font_color=STACK_COLOR, number_of_bars=7,root=canvas.tk,on_top=False)
canvas.add_sub_plot(stacked)
```

```python
# adding the time indication
time = Date.date(df=df, canvas=canvas.canvas, time_indicator="year", x_pos=1210, y_pos=865, height=45, width=0, back_ground_color=BAR_COLOR, font_color=FONT_COLOR)
canvas.add_sub_plot(time)
```

```python
# adding the total
t = Total.total(canvas=canvas, df=df, x_pos=1095, y_pos=135, height=700, width=200, font_color = FONT_COLOR, prefix = " ", unit = " $" )
canvas.add_sub_plot(t)
```

```python
# adding a sub title
sub_title = StaticText.static_text(canvas=canvas.canvas, text="GDP", x_pos=1140, y_pos=50, height=30, width=0, font_color=FONT_COLOR)
canvas.add_sub_plot(sub_title)
```

```python
# adding a sub title
sub_title = StaticText.static_text(canvas=canvas.canvas, text="per year", x_pos=1140, y_pos=82, height=30, width=0, font_color=FONT_COLOR)
canvas.add_sub_plot(sub_title)
```

```python
# adding a sub title
sub_title = StaticText.static_text(canvas=canvas.canvas, text="year", x_pos=1265, y_pos=835, height=30, width=0, font_color=FONT_COLOR)
canvas.add_sub_plot(sub_title)
```

```python
# adding a layer bar to the canvas
img = StaticImage.static_image(canvas=canvas.canvas, file="assets/Africa.png",x_pos=1160, y_pos=615, height=120, width=120,root=canvas.tk,on_top=True)
canvas.add_sub_plot(img)
```

```python
# adding a layer bar to the canvas
img = StaticImage.static_image(canvas=canvas.canvas, file="assets/GDPP.png",x_pos=405, y_pos=-9, height=1100, width=1050,root=canvas.tk,on_top=False)
canvas.add_sub_plot(img)
```

```python
# starting the animation
canvas.play(fps=FPS)
```
