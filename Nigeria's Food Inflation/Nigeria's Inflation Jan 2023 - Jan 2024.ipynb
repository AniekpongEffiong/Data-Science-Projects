```python
# importing all modules
from sjvisualizer import DataHandler
from sjvisualizer import Canvas
from sjvisualizer import StackedBarChart
from sjvisualizer import RecreateBarRace
from sjvisualizer import PieRace
from sjvisualizer import StaticText
from sjvisualizer import Date
from sjvisualizer import Total
from sjvisualizer import StaticImage
import json
```

```python
# define constants
FPS = 60
DURATION = 0.40
BACKGROUND_COLOR = (250, 250, 250)
TOTAL_COLOR = (0,0,0)
FONT_COLOR = (14,45,101)
BAR_COLOR = (255,255,255)
STACK_COLOR = (250,250,250)
DATE_COLOR = (39, 171, 95)
GREY_COLOR = (163, 167, 181)
number_of_frames=FPS*DURATION*60
```

```python
# load the data from the excel file
df = DataHandler.DataHandler(excel_file="data/Ascending.xlsx", number_of_frames=number_of_frames).df
```

```python
# load the data from the excel file
df1 = DataHandler.DataHandler(excel_file="data/Stacked.xlsx", number_of_frames=number_of_frames).df
```

```python
# load the data from the excel file
df2 = DataHandler.DataHandler(excel_file="data/Total.xlsx", number_of_frames=number_of_frames).df
```

```python
# loading colors
with open("colors.json") as f:
    colors = json.load(f)
```

```python
# creating the canvas to which we want to draw all elements
canvas = Canvas.canvas(bg=BACKGROUND_COLOR)
```

```python
# drawing a rectangle around working area
canvas.canvas.create_rectangle(-1,-1,1771,1771)
```

```python
# creating the bar race
bar_race = RecreateBarRace.bar_race(canvas=canvas.canvas, df=df,colors=colors, width=517, height=280, x_pos=697, y_pos=300, shift=700, decimal_places = 2, unit ="%", font_color=DATE_COLOR, number_of_bars=5)
canvas.add_sub_plot(bar_race)
```

```python
# creating the bar race
bar_race = RecreatedBarRace.bar_race(canvas=canvas.canvas, df=df, colors=colors, width=509, height=280, x_pos=702, y_pos=620, shift=700, decimal_places = 2, unit ="%", font_color=DATE_COLOR, number_of_bars=5)
canvas.add_sub_plot(bar_race)
```

```python
# creating the stacked bar chart
stacked = StackedBarChart.stacked_bar_chart(canvas=canvas.canvas, df=df1, colors=colors, x_pos=1156, y_pos=919, width=140, height=90, font_color=STACK_COLOR, number_of_bars=5.5,root=canvas.tk,on_top=False)
canvas.add_sub_plot(stacked)
```

```python
# adding the time indication
time = Date.date(df=df, canvas=canvas.canvas, time_indicator="month", x_pos=1220, y_pos=195, height=30, width=0, font_color=DATE_COLOR)
canvas.add_sub_plot(time)
```

```python
# adding the total
t = Total.total(canvas=canvas, df=df2, x_pos=752, y_pos=211, height=650, width=200, decimal_places=2, font_color = DATE_COLOR, prefix = "", unit = "%" )
canvas.add_sub_plot(t)
```

```python
# adding a layer bar to the canvas
img = StaticImage.static_image(canvas=canvas.canvas, file="assets/Inflation.png",x_pos=405, y_pos=5, height=1070, width=1100,root=canvas.tk,on_top=False)
canvas.add_sub_plot(img)
```

```python
# starting the animation
canvas.play(fps=FPS)
```
